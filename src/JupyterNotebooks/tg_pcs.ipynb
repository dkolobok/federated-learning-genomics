```python
import plotly.graph_objects as go

def add_pc_mean_std(go_fig, sscore, offset, name):
    y = sscore.describe().loc['mean', 'PC1_AVG': 'PC20_AVG']
    x = numpy.arange(1, 21) + offset

    std = sscore.describe().loc['std', 'PC1_AVG': 'PC20_AVG']

    go_fig.add_trace(go.Scatter(
            x=x,
            y=y,
            name=name,
            error_y=dict(
                type='data', # value of error bar given in data coordinates
                array=std,
                visible=True)
    ))

fig = go.Figure()
add_pc_mean_std(fig, train_sscore, 0.0, 'train mean and std')
add_pc_mean_std(fig, val_sscore, 0.2, 'val mean and std')
add_pc_mean_std(fig, test_sscore, -0.2, 'test mean and std')

fig.show()
```
