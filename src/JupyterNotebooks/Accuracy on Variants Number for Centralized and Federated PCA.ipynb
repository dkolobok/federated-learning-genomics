```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
centralized_data = pd.read_csv('/home/genxadmin/centralized-pca-pruning-accuracy.log', sep=' ', skiprows=1, header=None)
centralized_data.columns = ['pruning_threshold', 'variants_number', 'train', 'validation', 'test']
numbers = centralized_data['pruning_threshold'].unique()

centralized_train = []
centralized_validation = []
centralized_test = []

for number in numbers:
    subset = centralized_data.loc[centralized_data['pruning_threshold'] == number]
    
    centralized_train.append(subset['train'].sum() / len(subset))
    centralized_validation.append(subset['validation'].sum() / len(subset))
    centralized_test.append(subset['test'].sum() / len(subset))

plt.plot(numbers, centralized_train)
plt.plot(numbers, centralized_validation)
plt.plot(numbers, centralized_test)
plt.xticks(numbers)

federated_data = pd.read_csv('/home/genxadmin/federated-pca-pruning-accuracy.log', sep=' ', skiprows=1, header=None)
federated_data.columns = ['pruning_threshold', 'variants_number', 'train', 'validation', 'test']
numbers = federated_data['pruning_threshold'].unique()

federated_train = []
federated_validation = []
federated_test = []

for number in numbers:
    subset = federated_data.loc[federated_data['pruning_threshold'] == number]
    
    federated_train.append(subset['train'].sum() / len(subset))
    federated_validation.append(subset['validation'].sum() / len(subset))
    federated_test.append(subset['test'].sum() / len(subset))

plt.plot(numbers, federated_train)
plt.plot(numbers, federated_validation)
plt.plot(numbers, federated_test)

plt.legend(['train', 'validation', 'test', 'fed-train', 'fed-validation', 'fed-test'])
plt.grid()
plt.show()
```
